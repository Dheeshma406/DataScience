```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,6,8,9,10]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Graph")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y,color='green',linestyle='--',marker='*')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
plt.Figure(figsize=(3,3))
plt.plot(x,y,color='green',linestyle='--',marker='*')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph')
plt.show()

```

```python
courses=['Maths','Physics','Malayalam']
students=[10,20,30]
plt.bar(courses,students,color=['blue','red','black'])
plt.xlabel('Courses')
plt.ylabel('No of students')
plt.title('Details')
plt.show

```

```python
sizes=[10,20,30]
label=['A','B','C']
plt.pie(sizes,labels=label,autopct="%1.1f%%")
plt.axis("equal")
plt.title('Piechart')
plt.show()
```

```python

sizes=[10,20,30]
category=['I','II','III']
colors=['#653700','#F97306','#029386']
plt.pie(sizes,labels=category,autopct="%1.1f%%")
plt.axis("equal")
plt.title("piechart")
```

```python
import seaborn as sns
sizes=[10,20,30]
category=['I','II','III']
colors=sns.color_palette("viridis",len(sizes))
plt.pie(sizes,colors=colors,labels=category,autopct="%1.1f%%")
plt.axis("equal")
plt.title("piechart")
```

```python
sizes=[10,20,30]
category=['A','B','C']
colors=sns.color_palette('plasma',len(sizes))
plt.pie(sizes,labels=category,colors=colors,autopct="%11f%%")
plt.axis('equal')
plt.title('piechart')
```

```python
x=[10,20,30,40]
y=['A','B','C','D']
plt.scatter(x,y,color=['black','green','red','blue'])
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Chart')
```

```python
x=[10,20,30,40,50]
y=['A','B','C','D','E']
plt.barh(x,y,color=['blue','green','red','purple','Yellow'])
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('chart')
```

```python
x=[10,20,30,40,40]
y=[50,5,70,22,19,90,2,65]
plt.hist(x,y,bins=5,color=['blue','green','red'],edgecolor='black')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Chart')
```
