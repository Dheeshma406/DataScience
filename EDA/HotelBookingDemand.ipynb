**7.Perform Outliers for Numeric Columns**

```python
Q1 = df['adr'].quantile(0.25)
Q3 = df['adr'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['adr'] < (Q1 - 1.5 * IQR)) | (df['adr'] > (Q3 + 1.5 * IQR))]
print(len(outliers))
```

**Observation:**

 * A total of 2,488 outliers were detected in the adr column using the IQR method.

 * These represent unusually high or low average daily rates compared to the normal booking 
 
   price range.

**8.Remove the Outliers(ADR)**

```python
Q1 = df['adr'].quantile(0.25)
Q3 = df['adr'].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
df = df[(df['adr'] >= lower_limit) & (df['adr'] <= upper_limit)]
df
```

**Observation:**

 *  After removing outliers in the adr column using the IQR method, 2,488 extreme values were 
 
    eliminated. The dataset size reduced from 87,370 to 84,882 rows and 31 columns, ensuring a 
   
    cleaner and more reliable dataset for further analysis.

**9.Plot ADR Distribution**

```python
plt.figure(figsize=(10,5))
plt.hist(df['adr'], bins=30, edgecolor='black')
plt.title('Distribution of Average Daily Rate (ADR)')
plt.xlabel('ADR per night')
plt.ylabel('Frequency')
plt.show()
```

**Observation:**

 * The x-axis (ADR per night) shows the range of average daily room rates.

 * The y-axis (Frequency) shows how many bookings fall within each rate range.

 * Most hotel bookings have ADR values around â‚¹70â€“â‚¹120 per night.

 * A few values near 0 may represent free or complimentary stays.

**10.ADR by HotelType**

```python
avg_adr = df.groupby('hotel')['adr'].mean().reset_index()

plt.figure(figsize=(6,4))
plt.bar(avg_adr['hotel'], avg_adr['adr'], color=['orange', 'skyblue'])
plt.title('Average ADR by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Average ADR')
plt.show()
```

**Observation:**

 * The bar chart compares the average daily rate (ADR) between Resort Hotels and City Hotels.

 * From the chart, it is observed that the City Hotel has a higher average ADR than the Resort
 
   Hotel, indicating that City Hotels generally charge more per night.

 * The difference in rates suggests that City Hotels may cater to business travelers or urban 
 
   tourists willing to pay higher prices, while Resort Hotels offer more variable and often 
  
   lower rates depending on the season and vacation demand.

**11.Analyze ADR by Month**

```python

adr_month = df.groupby('arrival_date_month')['adr'].mean()

print(adr_month)

```

**Observation:**

 * The Average Daily Rate (ADR) is highest during August and lowest during January.
 
 * This shows that hotel prices increase during summer months (Juneâ€“August) due to high demand 
  
    and decrease during winter months (Novemberâ€“February), indicating an off-season period.

**Visualize ADR by Month**

```python
plt.bar(adr_month.index,adr_month.values)
plt.title('Average Daily Rate by Month')
plt.xlabel('Month')
plt.ylabel('ADR')
plt.xticks(rotation=45)
plt.show()
```
