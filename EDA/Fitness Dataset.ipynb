**12.Perform outlier detection on numeric features.**

```python
numeric_col=df.select_dtypes(include=['int64','float64']).columns
for cols in numeric_col:
  
  Q1=df[cols].quantile(0.25)
  Q3=df[cols].quantile(0.75)
  IQR=Q3-Q1
  outliers = df[(df[cols] < Q1-1.5*IQR) | (df[cols] > Q3+1.5*IQR)]
  print(f"Column: {cols}")
  print(f"Outlier count: {len(outliers)}")
  print(outliers[cols].head())
```

**13.Remove or treat the outliers appropriately**

```python
Q1=df[cols].quantile(0.25)
Q3=df[cols].quantile(0.75)
IQR=Q3-Q1
filtered_df=df[(df[cols]>=Q1-1.5*IQR)&(df[cols]<=Q3+1.5*IQR)]
print(filtered_df.shape)
```
