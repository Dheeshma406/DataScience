**12.Extract the genres(note:some movies may have multiple genres,split them.)**

```python

from collections import Counter
import ast

# function to extract genre names from the JSON-like strings,handling potential errors
 
def extract_genre_names(genres_str):
    try:

         genres=ast.literal_eval(genres_str)
         return[genre['name']for genre in genres]
    except(SyntaxError,ValueError):
         return['Unknown']        # Replace with a default value or handle error as needed.
    
    # Apply the function to the 'genres'column

df['genres_list']=df['genres'].apply(extract_genre_names)
    
    # Flatten the list of genres and count occurences

all_genres=[genre for sublist in df['genres_list']for genre in sublist]
genre_counts=Counter(all_genres)

    # Convert the counter to a dataframe for easy plotting

genre_df=pd.DataFrame(genre_counts.items(), columns=['Genre','Count'])
print("Genre Counts:")
for genre,count in genre_counts.items():
     print(f"{genre}:{count}")
    
    
```

**Observation:**
 
  * The output shows each movie genre and how many movies belong to it.

  * It clearly highlights which genres have the most film and which have fewer.

  * Movies with more than one genre are counted in each of their genres, and any missing or 
  
     invalid data appears as “Unknown”.   

**13.Count the number of movies in each genre.**


```python
df['genres_list']=df['genres'].apply(lambda x: extract_genre_names(x))
df=df.explode('genres_list')
genre_counts=df['genres_list'].value_counts()
print(genre_counts)
```
